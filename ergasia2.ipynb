Ερώτημα 1 - Logistic regression & Overfitting

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x,w):
    return ( 1 / ( 1 + math.exp( -w*x ) ) )
```

```python
x = np.arange(-10., 10., 0.01)
sigmoid_vectorized = np.vectorize(sigmoid)

sigmoid_w_1 = sigmoid_vectorized(x,1)
plt.plot(x,sigmoid_w_1)
plt.show()
```

```python
sigmoid_w_5 = sigmoid_vectorized(x,5)
plt.plot(x,sigmoid_w_5)
plt.show()
```

```python
try:
    sigmoid_w_100 = sigmoid_vectorized(x,100)
    plt.plot(x,sigmoid_w_100)
    plt.show()
except Exception as e:
    print(e)     
```

<b>Παρατηρήσεις - Απαντήσεις - Ερωτήματος 1</b><br />
Η λογιστική παλινδρόμηση (logistic regression) εμφανίζει συμπεριφορά overfitting για μεγάλες τιμές της παραμέτρου w: 

Το μοντέλο προσπαθεί να προσεγγίσει το 0 και το 1 αντίστοιχα όσο το δυνατόν πλησιέστερα και για αυτό οι τιμές των weights, δηλαδή της παραμέτρου w γίνονται ολοένα και μεγαλύτερες. Υποθετικά αν το w μπορούσε να πάρει ως τιμή το "άπειρο" το μοντέλο θα μπορούσε να πάρει τις τιμές 0 και 1, ενώ τώρα τις προσεγγίζει ασυμπτωτικά. Τα νούμερα γίνονται υπερβολικά μεγάλα και για αυτό προκύπτει το overfitting.

Εφαρμόζοντας regularization function θα μειωθεί το σφάλμα γενίκευσης αλλά όχι το σφάλμα εκπαίδευσης, γεγονός που σημαίνει ότι θα αποφύγουμε overfitting errors.

Θα χρησιμοποιούσα την L1 regularization function, καθώς χρησιμοποιεί τις ευκλείδιες αποστάσεις που προτιμάται από την L2 που χρησιμοποιεί διαφορά τετραγώνων. Εάν χρησιμοποιήσουμε την κανονικοποίηση L1 στην Logistic Regression, όλα τα λιγότερο σημαντικά χαρακτηριστικά θα μηδενιστούν, "εξαφανίζοντας" το overfitting.
